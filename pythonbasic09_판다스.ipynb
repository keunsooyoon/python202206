```python
# 분석하기 좋은 데이터

# 분석하기 좋은 데이터는 데이터 집합을 분석하기 좋은 상태로 만들어 놓은 것을 말한다. 
# 데이터 분석에서는 데이터 정리가 아주 중요하다. 실제로 데이터 분석 작업의 70~80% 를 
# 차지하는 작업이 데이터 정리 작업이다. (전처리 preprocessing)

# 깔끔한 데이터 Tidy Data
# - 데이터 분석 목적에 맞게 데이터를 수집하여 표(데이터프레임)로 만들어야 한다.  
# - 값은 행으로 구성한다. 
# - 변수는 열로 구성한다. 
```

```python
# 데이터 연결

import pandas as pd

df1 = pd.read_csv('concat_1.csv')
df2 = pd.read_csv('concat_2.csv')
df3 = pd.read_csv('concat_3.csv')

#  판다스로 엑셀이나 csv를 가져오면 데이터프레임으로 만들어서 가져온다. 

type(df1)
```

```python
# 가져온 세개의 데이터프레임을 하나로 연결한다. 

row_concat = pd.concat([df1,df2,df3])
row_concat
```

```python
new_row_series = pd.Series(['n1','n2','n3','n4'])

# 시리즈와 데이터프레임을 연결

pd.concat([df1, new_row_series])

# 데이터를 전처리하다 보면 의도하지 않게 결측처리되는 데이터가 빈번하게 발생한다. 
```

```python
i_row_concat = pd.concat([df1,df2,df3], ignore_index=True)
i_row_concat

# 기존 인덱스 체계는 삭제되고 새로운 인덱스가 전체에 부여된다. 
```

```python
# 열방향으로 데이터 연결하기
col_concat = pd.concat([df1,df2,df3], axis = 1) #생략하면 행방향이고 axis = 1은 열방향
col_concat
```

```python
col_concat['A']
```

```python
col_concat['F'] = ['1','2','3','4']
col_concat
```

```python
# Merge  매서드 로 연결하기

person = pd.read_csv('survey_person.csv')
site = pd.read_csv('survey_site.csv')
survey = pd.read_csv('survey_survey.csv')
visited = pd.read_csv('survey_visited.csv')
```

```python
visited_subset = visited.loc[[0,2,6],]
visited_subset
```

```python
# merge() 매서드는 기본적으로 앞에 적은 site 데이터프레임을 왼쪽으로 지정했고
# 첫 번째 이자값으로 지정한 데이터프레임 visited_subset 을 오른쪽으로 지정한다. 
# left_on, right_on 인자는 값이 일치해야할 왼쪽과 오른쪽 데이터프레임의 열을 지정한다. 
# 즉, 왼쪽 데이터프레임 site 의 열 name 과 오른쪽 데이터 프레임 visited_subset 의 열 site의
# 값이 일치하면 왼쪽 데이트 프레임을 기준으로 연결한다. 

o2o_merge = site.merge(visited_subset, left_on='name', right_on='site')
o2o_merge
```

```python
# 결측치 처리 (*****)

# 분석할 데이터에 결측치는 대부분 존재한다. 분석이나 머신러닝 전에 결측치를
# 처리하는 것은 매우 중요하다. 

# 결측치는 NaN, NAN, nan 로 표기할 수 있다. 

# 먼저 결측치 사용을 위해 numpy 를 불러오자

from numpy import NaN, NAN, nan

# 결측치는 0, '' 즉 영과 빈칸과는 다른것 이다. 결측치는 데이터 자체가 아예 없다는 의미이다. 
# 따라서 같다 라는 개념도 없다. 

print(NaN == True)
print(NaN == False)
print(NaN == '')
print(NaN == 0)
print(NaN == NaN)
print(NaN == NAN)
print(NaN == nan)
```

```python
# 그러면 결측치를 어떻게 확인할 수 있을까??

print(pd.isnull(NaN))
print(pd.isnull(NAN))
print(pd.isnull(nan))
```

```python
print(pd.notnull(NaN))
print(pd.notnull(NAN))
print(pd.notnull(nan))
```
